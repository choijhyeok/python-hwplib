```python
import requests

url = "http://localhost:7860/extract-text"
file_path = "./디지털정부혁신 추진계획.hwp" 

with open(file_path, 'rb') as f:
    files = {'file': (file_path, f)}
    response = requests.post(url, files=files)

response.json()
```
